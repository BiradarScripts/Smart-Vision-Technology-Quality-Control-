```python
import requests

# Replace this with your ngrok URL
ngrok_url = "https://1459-34-16-168-208.ngrok-free.app/webhook"

# Path to your image file
image_path = "/content/lays.jpg"  # Update this to the actual path of your image

# Open the image file in binary mode
with open(image_path, 'rb') as image_file:
    files = {
        'image': image_file,  # The key here should match what your server expects
    }
    # Add additional form data
    data = {
        "key": "Front packet",
        "another_key": "dsfgs"
    }

    # Send a POST request with both form data and the image file
    response = requests.post(ngrok_url, data=data, files=files)

# Print the response from the server
print("Response:", response.text)
```
