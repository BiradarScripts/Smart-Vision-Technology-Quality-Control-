```python
# Load model directly
from transformers import AutoModel,AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("stepfun-ai/GOT-OCR2_0", trust_remote_code=True)
model = AutoModel.from_pretrained("stepfun-ai/GOT-OCR2_0", trust_remote_code=True,low_cpu_mem_usage=True, device_map='cuda', use_safetensors=True, pad_token_id=tokenizer.eos_token_id)
model = model.eval().cuda()
```

```python
import pandas as pd
df =  pd.read_csv('/content/BigBasket Products.csv')
```

```python
df = df[['brand','product','type']]
```

```python
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df.drop_duplicates(inplace=True)
```

```python
def split_sentence(sentence):
    return sentence.split()
```

```python
image_file = '/content/cooking-oil.jpg'
res = model.chat_crop(tokenizer, image_file, ocr_type='ocr')
print(res)
```

```python
output = split_sentence(res)
output = list(set(output))
output = [x.lower() for x in output]
print(output)
```

```python
brand = df['brand'].drop_duplicates()
```

```python
max = 0
name=""
for i in brand:
    c = 0
    for j in output:
        if(str(j) in str(i)):
            c=c+1
    if(max<c):
        max = c
        name = i

print(name)
output.append(name)
```

```python
def canws(strings):
    a = False
    n = False
    for s in strings:
        if(s.isalpha()):
            a = True
        elif(s.isnumeric()):
            n = True
    return n
```
